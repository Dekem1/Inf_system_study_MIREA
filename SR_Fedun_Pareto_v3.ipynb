```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
DATA_URL = "https://raw.githubusercontent.com/sudakov/ai-lab/main/"

data = pd.read_csv(DATA_URL + "data.csv",header=None)
```

```python
data
```

```python
data_np=data.to_numpy()
```

```python
# ПРОВЕРКА СТРОК
def CHEK_NA_Dom_PO_PORETO(str_A, str_B):
    if (str_A[0] >= str_B[0]) and (str_A[1] >= str_B[1]) and (str_A[2] >= str_B[2]) and (str_A[3] >= str_B[3]) and (str_A[4] >= str_B[4]):       #Т.е идет проверка на то что столбцы строки А
        if (str_A[0] > str_B[0]) or (str_A[1] > str_B[1]) or (str_A[2] > str_B[2])  or (str_A[3] > str_B[3]) or (str_A[4] > str_B[4]):           #больше или равны столбцов строки B
            return 'str_A > str_B'
        else:
            return 'str_A = str_B'
    elif (str_A[0] <= str_B[0]) and (str_A[1] <= str_B[1]) and (str_A[2] <= str_B[2]) and (str_A[3] <= str_B[3]) and (str_A[4] <= str_B[4]):     # Аналогично, только для B
        if (str_A[0] < str_B[0]) or (str_A[1] < str_B[1]) or (str_A[2] < str_B[2]) or (str_A[3] < str_B[3]) or (str_A[4] < str_B[4]):
            return 'str_A < str_B'
        else:
            return 'str_A = str_B'
    else:
        return 'str_A = str_B'
```

```python
result = np.array([1])
upper = []
for i in range(len(data_np)-1):
    new_row = np.array([-1])
    result = np.vstack([result, new_row])
for i in range(len(data_np)-1):
    for j in range(i+1, len(data_np)):
        comp = CHEK_NA_Dom_PO_PORETO(data_np[i], data_np[j])
        if comp == 'str_A = str_B':
            if result[i] == -1:
                result[i] = 1
            if result[j] == -1:
                result[j] = 1
        if comp == 'str_A > str_B':
            if result[i] == -1:
                result[i] = 1
            result[j] = 0
        if comp == 'str_A < str_B':
            result[i] = 0
            if result[j] == -1:
                result[j] = 1
print(result)
```
